```python
import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image
```

```python
img = Image.open('skater.jpg')
_ = plt.imshow(img)
```

```python
imggray = img.convert('LA')
_ = plt.imshow(imggray)
```

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
_ = plt.imshow(imgmat, cmap='gray')
```

```python
U, sigma, V = np.linalg.svd(imgmat) # eignevalues arranged in descending order in dia(sigmma)
# singular values are arranged in descending order in D, first left left-singular vector of U 
# and first right-singular vector of V, most prominent fearure of image 
```

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
_ = plt.imshow(reconstimg)
```

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
_ = plt.imshow(reconstimg, cmap='gray')
```

```python
for i in [2,4,8,16,32,64]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg)
    title = "n = %s"
    plt.title(title)
    plt.show()
```
